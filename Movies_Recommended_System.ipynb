```python
movies_data= movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
# Drop down the null values
movies_data = movies_data.dropna()
```

```python
#'[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'
# to change the format ['Action','Adventure','Fantasy','Science Fiction']
```

```python
# In the columns('genres', 'keywords') many names 1 dictionary so we can only 3 items taken in dictionary and covert to list
import ast
def convert(obj):
    l = []
    for i in ast.literal_eval(obj):
        l.append(i['name'])
                 
    return l
        
```

```python
# copy the dataset after preprocessing
movies_data = movies_data.copy()
movies_data['genres'] = movies_data['genres'].apply(convert)
movies_data['keywords'] = movies_data['keywords'].apply(convert)
```

```python
movies_data.head(1)
```

```python
# In the columns('cast') many names 1 dictionary so we can only 3 items taken in dictionary and covert to list

import ast
def convert3(obj):
    l = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter !=3:
            l.append(i['name'])
            counter+=1
        else :
            break
                 
    return l
        
```

```python
movies_data['cast'] = movies_data['cast'].apply(convert3)
```

```python
movies_data['cast']
```

```python
# In this column('crew') movies of Directors names you can arrange a list  
import ast
def fect_director(obj):
    l = []
    for i in ast.literal_eval(obj):
       if i['job'] == 'Director':
            l.append(i['name'])
            break

    return l
        
```

```python
movies_data['crew'] = movies_data['crew'].apply(fect_director)
```

```python
movies_data.head()
```

```python
# here column('overview') is 'string' so to convert a list 
movies_data['overview'] = movies_data['overview'].apply(lambda x: x.split())
```

```python
movies_data.head()
```

```python
# 'Sam Worthington' to remove the sapce 'SamWorthington' in every columns space remove
movies_data['genres'] = movies_data['genres'].apply(lambda x :[i.replace(" ","") for i in x])
movies_data['keywords'] = movies_data['keywords'].apply(lambda x :[i.replace(" ","") for i in x])
movies_data['cast'] = movies_data['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies_data['crew'] = movies_data['crew'].apply(lambda x :[i.replace(" ","") for i in x])
```

```python
movies_data.head()
```

```python
# concat of the columns
movies_data['tags'] = movies_data['overview'] + movies_data['genres'] + movies_data['keywords'] + movies_data['cast'] + movies_data['crew']
```

```python
movies_data.head()
```

```python
# now creat a new dataframe
new_df = movies_data[['movie_id','title','tags']]
```

```python
new_df.head()
```

```python
# here column('tags') a list to covert  string
new_df = new_df.copy()
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
new_df['tags'][0]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
# NLP nltk library
import nltk
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)


```

```python
# Add to stem in apply tags
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
# feature  CountVectorizer

from sklearn.feature_extraction.text import CountVectorizer

cv= CountVectorizer(max_features=5000,stop_words='english')


# Fit the text
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors[0]
```

```python
# To find the Euclidean Distance but it is not relable in heigh dimensional so we can fin the Cosine_Distance
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vectors)
```

```python
#similarity[0]
# to change the index and to recommended top 5 movies in related movie
sorted(list(enumerate(similarity[0])), reverse=True,key=lambda x:x[1])[1:6]
```

```python
# to change the index and to recommended top 5 movies in related movie
def recommend(movie):
    movies_index = new_df[new_df['title'] == movie].index[0]
    distance = similarity[movies_index]
    movies_list = sorted(list(enumerate(distance)),reverse = True, key = lambda x:x[1])[1:6]
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
       
     
```

```python
#recommend('Man of Steel')
#print()
recommend('Avatar')
```
